```python
df = pd.read_parquet('df_checkpoint.parquet')
df = df[df['tweet_created_at'] >= '2018-01-01']
```

```python
# initialize class
#stdc = STDC(num_rows=100000,dimensions=2, time_type='intrinsic', timeframe=1000)
stdc = STDC(raw_data=df.head(10000000), field_names=['author_id', 'retweeter_id', 'tweet_created_at'],dimensions=2, time_type='actual', timeframe="%Y-%m")
```

```python
stdc.calculate_timeframe()
```

```python
stdc.raw_data
```

```python
stdc.calculate_biadjacency_matrix()
```

```python
stdc.calculate_positions()
```

```python
stdc.calculate_reduced_positions()
```

```python
stdc.calculate_velocities()
```

```python
stdc.calculate_aligned_reduced_positions()
```

```python
stdc.velocities
```

```python
v_stats = stdc.velocities.groupby(['t1','t2']).agg(['mean','var'])
v_stats
```

```python
p_stats = stdc.aligned_reduced_positions.groupby(['t1','t2']).agg(['mean','var'])
```

```python
plt.plot(p_stats[0]['mean'], p_stats[1]['mean'], marker='o', label='Trajectory')

plt.quiver(
    p_stats[0]['mean'][:-1], p_stats[1]['mean'][:-1],                # start points
    p_stats[0]['mean'].diff()[1:], p_stats[1]['mean'].diff()[1:],    # vector components
    scale_units='xy', angles='xy', scale=1, color='r', width=0.006
)
```

```python
vol_ts = np.sqrt(p_stats.xs('var', axis=1, level=1)).prod(axis=1)
```

```python
temp_ts = v_stats.xs('var', axis=1, level=1).sum(axis=1)
```

```python
vcom_ts = np.sqrt(np.power(v_stats.xs('mean', axis=1, level=1), 2).sum(axis=1)) # V = (V_x, V_y, ...) -> |V| = sqrt(V_x^2 + V_y^2 + ...)
```

```python
stats = pd.DataFrame({'Vol': vol_ts, 'Temp': temp_ts, 'V_CoM': vcom_ts})
stats
```
