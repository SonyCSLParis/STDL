```python
from STDC import STDC
```

```python
# initialize class
stdc = STDC(timeframe='Y')

# store data
data = stdc.data()
```

```python
# view data
stdc.view()
```

```python
stdc.calculate_positions(data, comparison='absolute')
```

```python
# biadjacency matrix
stdc.compute_biadjacency_matrix(raw_data=data, leader_col=stdc.leader_col, follower_col=stdc.follower_col)
```

```python
# get relative matrix
stdc.get_relative_matrix()
```

```python
# get absolute matrix
stdc.get_absolute_matrix()
```

```python
# unique timeframes
stdc.get_unique_timeframes(data)
```

```python
# unique entities
stdc.get_unique_entities(data)
```
