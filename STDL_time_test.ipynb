```python
from STDC import STDC
import time

# methods to test in dependency order
methods_to_test = [
    "calculate_timeframe",
    "calculate_biadjacency_matrix",
    "calculate_unreduced_positions",
    "calculate_positions",
    "calculate_velocities"
]

sizes = [1_000, 5_000, 10_000, 50_000, 100_000]
results = []

for n in sizes:
    print(f"\nTesting with {n} rows")
    s = STDC(num_rows=n, dimensions=2)

    for method in methods_to_test:
        func = getattr(s, method)
        start = time.perf_counter()
        try:
            func()
        except Exception as e:
            elapsed = None
            print(f"{method} failed with {e}")
        else:
            elapsed = time.perf_counter() - start
            print(f"{method}: {elapsed:.4f} sec")
```
